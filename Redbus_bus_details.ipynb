# Andhra 

```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 
import time
import pandas as pd

df=pd.read_csv(r"Bus_Routes_links/df_Andhra.csv")
# Lists to store bus details
Bus_names_andhra = []
Bus_types_andhra = []
Start_Time_andhra = []
End_Time_andhra = []
Ratings_andhra = []
Total_Duration_andhra = []
Prices_andhra = []
Seats_Available_andhra = []
Route_names = []
Route_links = []

# Loop through each link in the DataFrame
for i, r in df.iterrows():
    driver_andhra=webdriver.Chrome()
    link = r["Route_links"]
    routes = r["Route_Name"]

    # Open the link
    driver_andhra.get(link)
    time.sleep(3)
    # Click on elements to reveal bus details
    elements = driver_andhra.find_elements(By.XPATH, f"//a[contains(@href, '{link}')]")
    for element in elements:
        element.click()
    time.sleep(10)
    driver_andhra.find_element(By.XPATH,"//span[@class='next']").click() # since the bus on the time we search varies.we can overcome by going to the following day to get all data
    time.sleep(10)
    


    # click elements to views bus
    try:
        driver_andhra.find_element(By.XPATH, "//div[@class='button']").click()
        time.sleep(3)
    except :
        print(f"Button not found for link: {link}")


    
    while True:
        old_page_source = driver_andhra.page_source

        # Scroll down to the bottom of the page using JavaScript
        driver_andhra.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Wait for new content to load

        new_page_source = driver_andhra.page_source

        if new_page_source == old_page_source:
            break




# Extract bus details
    bus_names = driver_andhra.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    bus_types = driver_andhra.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    start_times = driver_andhra.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
    end_times = driver_andhra.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
    total_durations = driver_andhra.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
    ratings = driver_andhra.find_elements(By.XPATH, "//div[@class='clearfix row-one ']/div[@class='column-six p-right-10 w-10 fl']/div[@class='rating-sec lh-24']")
    prices = driver_andhra.find_elements(By.XPATH, '//*[@class="fare d-block"]')
    seats_available = driver_andhra.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")

    # Print the extracted information
    for i in range(len(bus_names)):
        Bus_names_andhra.append(bus_names[i].text)
        Bus_types_andhra.append(bus_types[i].text)
        Start_Time_andhra.append(start_times[i].text)
        End_Time_andhra.append(end_times[i].text)
        Total_Duration_andhra.append(total_durations[i].text)
        Prices_andhra.append(prices[i].text)
        Seats_Available_andhra.append(seats_available[i].text)
        Route_names.append(routes)
        Route_links.append(link)
        if i < len(ratings):
            Ratings_andhra.append(ratings[i].text)
        else:
            Ratings_andhra.append("Nil")

    # Close the WebDriver
    driver_andhra.quit()

data = {
    'Bus_name': Bus_names_andhra,
    'Bus_type': Bus_types_andhra,
    'Start_time': Start_Time_andhra,
    'End_time': End_Time_andhra,
    'Total_duration': Total_Duration_andhra,
    'Price': Prices_andhra,
    "Seats_Available":Seats_Available_andhra,
    "Ratings":Ratings_andhra,
    'Route_link': Route_links,
    'Route_name': Route_names
}

print(len(Bus_names_andhra),len(Bus_types_andhra),len(Start_Time_andhra),len(End_Time_andhra),len(Total_Duration_andhra),len(Route_links))
df_bus_andhra = pd.DataFrame(data)
df_bus_andhra.to_csv(r"C:\Users\LENOVO\Desktop\GUVI\Redbus\Bus_details\df_andhra.csv",index=False)
```

```python
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
import datetime
from datetime import datetime, timedelta
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

df = pd.read_csv(r"C:\Users\LENOVO\Desktop\GUVI\Redbus\Bus_Routes_links\df_Meghalaya.csv")
# Initialize the WebDriver


# Lists to store bus details
Bus_names_Meghalaya = []
Bus_types_Meghalaya = []
Start_Time_Meghalaya = []
End_Time_Meghalaya = []
Ratings_Meghalaya = []
Total_Duration_Meghalaya = []
Prices_Meghalaya = []
Seats_Available_Meghalaya = []
Route_names = []
Route_links = []

# Loop through each link in the DataFrame
for i, r in df.iterrows():
    driver_Meghalaya=webdriver.Chrome()
    link = r["Route_links"]
    routes = r["Route_Name"]

    # Open the link
    driver_Meghalaya.get(link)
    # Click on elements to reveal bus details
    time.sleep(3)
    elements = driver_Meghalaya.find_elements(By.XPATH, f"//a[contains(@href, '{link}')]")
    for element in elements:
        element.click()
    time.sleep(10)
    driver_Meghalaya.find_element(By.XPATH,"//span[@class='next']").click()
    time.sleep(10)
    


    # click elements to views bus
    try:
        driver_Meghalaya.find_element(By.XPATH, "//div[@class='button']").click()
        time.sleep(3)
    except :
        print(f"Button not found for link: {link}")


    
    while True:
        old_page_source = driver_Meghalaya.page_source

        # Scroll down to the bottom of the page using JavaScript
        driver_Meghalaya.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Wait for new content to load

        new_page_source = driver_Meghalaya.page_source

        if new_page_source == old_page_source:
            break




# Extract bus details
    bus_names = driver_Meghalaya.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    bus_types = driver_Meghalaya.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    start_times = driver_Meghalaya.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
    end_times = driver_Meghalaya.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
    total_durations = driver_Meghalaya.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
    ratings = driver_Meghalaya.find_elements(By.XPATH, "//div[@class='clearfix row-one ']/div[@class='column-six p-right-10 w-10 fl']/div[@class='rating-sec lh-24']")
    prices = driver_Meghalaya.find_elements(By.XPATH, '//*[@class="fare d-block"]')
    seats_available = driver_Meghalaya.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")

    # Print the extracted information
    for i in range(len(bus_names)):
        Bus_names_Meghalaya.append(bus_names[i].text)
        Bus_types_Meghalaya.append(bus_types[i].text)
        Start_Time_Meghalaya.append(start_times[i].text)
        End_Time_Meghalaya.append(end_times[i].text)
        Total_Duration_Meghalaya.append(total_durations[i].text)
        Prices_Meghalaya.append(prices[i].text)
        Seats_Available_Meghalaya.append(seats_available[i].text)
        Route_names.append(routes)
        Route_links.append(link)
        if i < len(ratings):
            Ratings_Meghalaya.append(ratings[i].text)
        else:
            Ratings_Meghalaya.append("Nil")   
    # Close the WebDriver
    driver_Meghalaya.quit()
data = {
    'Bus_name': Bus_names_Meghalaya,
    'Bus_type': Bus_types_Meghalaya,
    'Start_time': Start_Time_Meghalaya,
    'End_time': End_Time_Meghalaya,
    'Total_duration': Total_Duration_Meghalaya,
    'Price': Prices_Meghalaya,
    "Seats_Available":Seats_Available_Meghalaya,
    "Ratings":Ratings_Meghalaya,
    'Route_link': Route_links,
    'Route_name': Route_names
}

df_bus_Meghalaya = pd.DataFrame(data)
df_bus_Meghalaya.to_csv(r"C:\Users\LENOVO\Desktop\GUVI\Redbus\Bus_details\df_Meghalaya.csv",index=False)
df_bus_Meghalaya
```
