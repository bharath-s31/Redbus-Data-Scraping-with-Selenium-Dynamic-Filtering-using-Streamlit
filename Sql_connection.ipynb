```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\LENOVO\Desktop\GUVI\Redbus\Bus_details\Redbus_final_data.csv",encoding='latin1')
df
```

```python
# Remove 'INR' and any extra spaces from Price column and change the dtype
df['Price'] = df['Price'].str.replace(r'\D','',regex=True).str.strip().astype(float)
print(df)
```

```python
# remove 'Nil' values and replace with 0 in Ratings column and change datatype
df['Ratings'] = df['Ratings'].str.replace('Nil','0').str.strip().astype(float)
print(df)
```

```python
df['Seats_Available']=df['Seats_Available'].str.replace(r'\D','',regex=True).str.strip().astype(int)
df['Start_time'] = pd.to_datetime(df['Start_time'], format='%H:%M', errors='coerce').dt.time
df['End_time']=pd.to_datetime(df['End_time'], format='%H:%M', errors='coerce').dt.time
df
```
